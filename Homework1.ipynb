```python
companies=["2222.SR","BRK-B", "AAPL", "MSFT", "GOOG", "JPM"]
list_of_sum_dividends=[]
for company in companies:
    data=yf.Ticker(company)
    actions_of_all_the_time=data.get_actions()
    data_2023=actions_of_all_the_time[actions_of_all_the_time.index.year==2023]
    list_of_sum_dividends.append(sum(data_2023.Dividends))


print(list_of_sum_dividends)
```

```python
list_of_primary_data=[sr_2222,brk_b,aapl,msft,goog,jpm]
brk_b.tail()
```

```python
brk_b["Adj Close"].iloc[-1]
```

```python
import numpy as np
# Extract the last value of Adj.Close
adj_closes =[]
for data in list_of_primary_data:
    adj_close=data["Adj Close"].iloc[-1]
    adj_closes.append(adj_close)


result=(np.array(list_of_sum_dividends)/np.array(adj_closes))*100
for number in result :
    print(round(number,2))
```
