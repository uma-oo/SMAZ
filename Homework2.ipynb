```python
# Customize the function for our use

def calculate_avg_price(price_range):
    if pd.isnull(price_range) or price_range == '':
        return np.nan
    elif '-' in price_range:
        prices = price_range.split(' - ')
        try:
            avg_price = (float(prices[0][1:]) + float(prices[1][1:])) / 2
            return avg_price
        except ValueError:
            return np.nan  # Handle non-numeric values in the price range
    else:
        try:
            return float(price_range[1:])
        except ValueError:
            return np.nan  # Handle non-numeric values in the price


examples = ["$8.00 - $10.00", "$10.00", np.nan, "-"]
for example in examples:
    res = calculate_avg_price(example)
    print(f"{example} ==> {res}")
```

```python
filings_data['Avg_price'] = filings_data["Price Range"].apply(lambda x:calculate_avg_price(x))
```

```python
filings_data.tail(10)
```

```python
# Create the "Shares_offered_value" which equals to "Shares Offered" * "Avg_price" (when both columns are defined; otherwise, it's NaN)

def calculate_product(value_1,value_2):
    if value_1==np.nan or value_2==np.nan:
        return np.nan
    return value_1*value_2

```

```python
filings_data['Shares_offered_value'] =  filings_data.apply(lambda row: calculate_product(row['Shares Offered'], row['Avg_price']), axis=1)
```
